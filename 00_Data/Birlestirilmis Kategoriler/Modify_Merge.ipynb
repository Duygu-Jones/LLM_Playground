```python
import pandas as pd     
```

```python
df_18 = pd.read_csv('df_yikama_18.csv')
df_19 = pd.read_csv('df_yikama_19.csv')
df_20 = pd.read_csv('df_yikama_20.csv')
df_21 = pd.read_csv('df_yikama_21.csv')
df_22 = pd.read_csv('df_yikama_22.csv')
df_23 = pd.read_csv('df_yikama_23.csv')
df_24 = pd.read_csv('df_yikama_24.csv')
```

```python
df_21 = df_21.rename(columns={"TESLİM \nSÜRESİ": "TESLİM TARİHİ", "TESLİM EDİLEN TARİH": "TESLİM \nSÜRESİ"})
```

```python
def remove_year_from_columns(df):
    df.columns = [col.split('/')[0].strip() if '/' in col else col for col in df.columns]
    return df
dataframes = [df_18, df_19, df_20, df_21, df_22, df_23, df_24]
dataframes_standardized = [remove_year_from_columns(df) for df in dataframes]
df_18, df_19, df_20, df_21, df_22, df_23, df_24 = dataframes_standardized
```

```python
dataframes = [df_18, df_19, df_20, df_21, df_22, df_23, df_24]  # Düzenlenmiş tüm DataFrame'ler
merged_df = pd.concat(dataframes, ignore_index=True)
print(merged_df.head())
```

```python
merged_df.to_csv("yikama.csv", index=False)
```
