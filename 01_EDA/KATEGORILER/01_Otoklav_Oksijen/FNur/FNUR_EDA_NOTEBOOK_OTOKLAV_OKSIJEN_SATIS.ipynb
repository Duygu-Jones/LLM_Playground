Metinler detaylı analiz edilmeli yazım yanlışları, ing - tr olarak yazılan ifadeler gibi.

```python
# 'siparisi_veren' sütununda temizlik ve standartlaştırma işlemleri
def standardize_column(df_kategori, column_name):
    # Tüm verileri büyük karakterlere çevirme
    df_kategori[column_name] = df_kategori[column_name].str.upper()
    
    # -, + gibi sembollerin temizlenmesi
    df_kategori[column_name] = df_kategori[column_name].str.replace('-', ' ')
    df_kategori[column_name] = df_kategori[column_name].str.replace(' +', ' ', regex=True).str.strip()
    
    # Sondaki noktaları temizleyen işlemler
    df_kategori[column_name] = df_kategori[column_name].str.rstrip('..')
    df_kategori[column_name] = df_kategori[column_name].str.rstrip('...')
    df_kategori[column_name] = df_kategori[column_name].str.rstrip('....')
    df_kategori[column_name] = df_kategori[column_name].str.rstrip('.....')
    
    # Aynı anlama gelen ifadelerin standartlaştırılması
    standard_replacements = {
        'DMO -BURSA BÖLGE MÜDÜRÜLÜĞÜ': 'DMO BURSA BÖLGE MÜDÜRLÜĞÜ',
        'ISTANBUL IL SAGLIK MUDURLUGU': 'İSTANBUL İL SAĞLIK MÜDÜRLÜĞÜ',
        'UNIVERSITY OF ANKARA': 'ANKARA ÜNİVERSİTESİ',
        'Turkey': 'Türkiye',
        'Istanbul': 'İstanbul',
        'DMO 1 NOLU SÜREÇLER DAİRE BAŞKANLIĞI': 'DMO 1 NOLU OPERASYONEL SÜREÇLER DAİRE BAŞKANLIĞI',
        'ERCİYES ÜNİVERSTESİ': 'ERCİYES ÜNİVERSİTESİ',
        'KIMA HEALT PARTNER DAKAR': 'KIMA HEALTH PARTNER DAKAR',
        'MODERN MEDICAL': 'MODERN MEDİKAL',
        'YENİ HAYAT TIP MERKEZİ ORJİN SAĞLIK HİZM': 'YENİ HAYAT TIP MERKEZİ',
        'MME,LLC': 'MME, LLC',
        'BİOKAN SAĞLIK ÜRÜNLERİ SAN. VE TİC. A.Ş': 'BİOKAN SAĞLIK ÜRÜNLERİ SAN. VE TİC. LTD. ŞTİ.',
        'LLC MK BIO': 'LLC MK BİO',
        'SIMAD UNİVERSİTESİ': 'SİMAD ÜNİVERSİTESİ',
        'RAWAAT ALMAJARRA COMPANY': 'RAWAT ALMAJARRA',
        'TME': 'TM'
    }
    
    # DMO genel müdürlükleri için düzenlemeler
    dmo_replacements = {
        'DMO GENEL MÜDÜRLÜK': 'DMO GENEL MÜDÜRLÜĞÜ',
        'DMO PAZARLAMA GENEL MÜDÜRLÜĞÜ': 'DMO GENEL MÜDÜRLÜĞÜ',
        "DMO 1 NOLU OPERASYONEL SÜREÇLER DAİRESİ BAŞKAN": "DMO 1 NOLU OPERASYONEL SÜREÇLER DAİRE BAŞKANLIĞI",
        "DMO 1 NOLU OPERASYONEL SÜREÇLER DAİRESİ BAŞKANLIĞI": "DMO 1 NOLU OPERASYONEL SÜREÇLER DAİRE BAŞKANLIĞI",
    }
    
    # NICU ile ilgili kayıtlar
    nicu_replacements = {
        'NI CONSULTİNG / MR. NİCU': 'ROMANYA',
        'NİCU': 'ROMANYA',
        'ROMANYA NICU': 'ROMANYA',
        'ROMANYA NİCU': 'ROMANYA',
        'NİCU ROMANYA': 'ROMANYA',
    }
    
    # Ülke isimleri ve özel isimlerin standartlaştırılması
    country_and_special_replacements = {
        'AZERBEYCAN': 'AZERBAYCAN',
        'RUSYA FEDARASYONU': 'RUSYA',
        'NİGERİA': 'NİJERYA',
        'LEBANON': 'LÜBNAN',
        'CHILE': 'ŞİLİ',
        'VİTTA': 'VİTTA MEDİCAL',
        'ERCİYES ÜNİVERSİTESİ İDARİ VE MALİ İŞLER BAŞKANLIĞI': 'ERCİYES ÜNİVERSİTESİ'
    }
    
    # MERİVAN ile ilgili isimler
    merivan_replacements = {
        "MERİVAN IRAK": "MERİVAN IRAK",
        "IRAK MERİVAN": "MERİVAN IRAK",
        "IRAK MERİVAN MODERN MEDİKAL": "MERİVAN IRAK",
        'MERİVAN': "MERİVAN IRAK"
    }
    
    # Tüm eşleşmeleri uygulama
    df_kategori[column_name] = df_kategori[column_name].replace(standard_replacements)
    df_kategori[column_name] = df_kategori[column_name].replace(dmo_replacements)
    df_kategori[column_name] = df_kategori[column_name].replace(nicu_replacements)
    df_kategori[column_name] = df_kategori[column_name].replace(country_and_special_replacements)
    df_kategori[column_name] = df_kategori[column_name].replace(merivan_replacements)
    
    # Kısaltmaları standart hale getirme
    df_kategori[column_name] = df_kategori[column_name].apply(lambda x: x.replace(" SAN ", " SAN. ") if isinstance(x, str) else x)
    df_kategori[column_name] = df_kategori[column_name].apply(lambda x: x.replace(" TİC ", " TİC. ") if isinstance(x, str) else x)
    df_kategori[column_name] = df_kategori[column_name].apply(lambda x: x.replace(" A Ş", " A.Ş.") if isinstance(x, str) else x)
    df_kategori[column_name] = df_kategori[column_name].apply(lambda x: x.replace(" LTD ", " LTD. ") if isinstance(x, str) else x)
    df_kategori[column_name] = df_kategori[column_name].apply(lambda x: x.replace(" ŞTİ", " ŞTİ.") if isinstance(x, str) else x)
    
    return df_kategori
```

```python
from rapidfuzz import fuzz
import pandas as pd

# %40 ile %99 arasında benzerlik olanları arama
similar_entries = []
for i, name1 in enumerate(df_kategori['siparisi_veren']):
    for j, name2 in enumerate(df_kategori['siparisi_veren']):
        if i < j:  # İki kez karşılaştırmamak için
             # DMO veya BÖLGE MÜDÜRLÜĞÜ geçen kayıtları atla
            if "DMO" in name1 or "BÖLGE MÜDÜRLÜĞÜ" in name1 or "DMO" in name2 or "BÖLGE MÜDÜRLÜĞÜ" in name2:
                continue
            similarity = fuzz.ratio(name1, name2)
            if 80 < similarity < 100:  # %40'tan büyük ve %100'den küçük benzerlik
                similar_entries.append((name1, name2, similarity))

# Sonuçları gösterme
similar_entries_df = pd.DataFrame(similar_entries, columns=["Name1", "Name2", "Similarity"])
similar_entries_df
```

```python
import re

def duzenle_urun_modeli(df):
    
     # Başındaki ve sonundaki boşlukları kaldırma
    df['urun_modeli'] = df['urun_modeli'].str.strip()
    
    # Tüm boşlukları "-" ile değiştirme ve büyük harf düzeni
    df['urun_modeli'] = df['urun_modeli'].str.replace(r'\s+', '-', regex=True).str.upper()
    
    # Spesifik değişiklikler
    df['urun_modeli'] = df['urun_modeli'].replace({
        'GOLDBERG-75-V': 'GOLDBERG-75-V',
        'GOLDBERG-75V': 'GOLDBERG-75-V',
        'GOLDBERG-160-S': 'GOLDBERG-160-S',
        'GOLDBERG-160-S-(40*40)': 'GOLDBERG-160-S',
        'GOLDBERG-840-D': 'GOLDBERG-840-D',
        'GOLDBERG-840-D-REVİZYON': 'GOLDBERG-840-D',
        'GOLBERG-840-D': 'GOLDBERG-840-D',
        'STR-4410-S-(ESKI-KAZAN)' : 'STR-4410-S',
        'STR-6613-D-(TERS)' : 'STR-6613-D',
        'GOLDBERG-840D' : 'GOLDBERG-840-D',
        'GOLBERG-422-D' : 'GOLDBERG-422-D',
        'ES-100' : 'GOLDBERG-100-V'
    })
    
    df['urun_modeli'] = df['urun_modeli'].str.replace('VEARAŞTIRMA', 'VE ARAŞTIRMA')
    
    # OXYGOLD olanları OXY-GOLD yapma
    df['urun_modeli'] = df['urun_modeli'].str.replace(r'\bOXYGOLD\b', 'OXY-GOLD', regex=True)
    df['urun_modeli'] = df['urun_modeli'].str.replace(r'\bOXYGOLD-', 'OXY-GOLD-', regex=True)
    
    
    return df
```
