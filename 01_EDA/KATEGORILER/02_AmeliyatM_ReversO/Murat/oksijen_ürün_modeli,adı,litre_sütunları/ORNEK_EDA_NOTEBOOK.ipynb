```python
# Ürün adını güncelleme fonksiyonu
def urun_adi_guncelle(row):
    model = row['urun_modeli']
    kategori = row['kategori']
    
    if pd.isna(model):
        return row['urun_adi']  # Model boşsa, bir değişiklik yapma
    
    # Sadece OKSIJEN kategorisine ait OXYGOLD veya OXY-GOLD ürünleri için güncelleme
    if kategori == "OKSIJEN" and ('OXYGOLD' in model or 'OXY-GOLD' in model):
        # Burada OXY-GOLD veya OXYGOLD için ekleme yapıyoruz
        urun_adi = f"{model} OKSİJEN ÜRETİM VE DEPOLAMA SİSTEMİ"
    else:
        urun_adi = row['urun_adi']  # Diğer durumlarda mevcut adı kullan

    return urun_adi

# 'urun_adi' sütununu güncelleme
df['urun_adi'] = df.apply(urun_adi_guncelle, axis=1)

# Ürün modelini güncelleme fonksiyonu
def urun_modeli_guncelle(model):
    if pd.isna(model):
        return np.nan  # Model boşsa, bir değişiklik yapma

    # Boşlukları '-' ile değiştirme
    model = model.replace(' ', '-').strip()

    return model

# 'urun_modeli' sütununu güncelleme
df['urun_modeli'] = df['urun_modeli'].apply(urun_modeli_guncelle)

# Litrelik değerleri düzenleme
df['litre'] = df['litre'].str.replace('LT', '').str.replace('lt', '').astype(str)

# Güncellenmiş DataFrame'i gösterme
print(df)
```

```python
import re
# Ürün model ve adını güncelleme fonksiyonu (yalnızca OKSIJEN kategorisinde)
def urun_guncelle(row):
    model = row['urun_modeli']
    urun_adi = row['urun_adi']

    # Model güncellemeleri için sözlük
    model_guncellemeleri = {
        'OXYGOLD-75': 'OXY-GOLD-75',
        'OXYGOLD-3': 'OXY-GOLD-3',
        'OXYGOLD-37': 'OXY-GOLD-37',
        'OXYGOLD-22': 'OXY-GOLD-22'
    }

    # Model güncelleme
    if pd.notna(model):  # Boş olmayan değerler için işlem yap
        yeni_model = model_guncellemeleri.get(model, model)
    else:
        yeni_model = model

    # Ürün adını güncelleme: Hem "OXYGOLD X" hem de "OXY-GOLD X" -> "OXY-GOLD-X"
    if isinstance(urun_adi, str):  # Sadece string değerler için işlem yap
        yeni_urun_adi = re.sub(r"OXYGOLD (\d+)", r"OXY-GOLD-\1", urun_adi)
        yeni_urun_adi = re.sub(r"OXY-GOLD (\d+)", r"OXY-GOLD-\1", yeni_urun_adi)
    else:
        yeni_urun_adi = urun_adi

    # Güncellenmiş değerleri geri döndür
    row['urun_modeli'] = yeni_model
    row['urun_adi'] = yeni_urun_adi
    return row

# Tüm DataFrame'de güncelleme yapma
df = df.apply(urun_guncelle, axis=1)

# Yalnızca OKSIJEN kategorisini filtreleme
oksijen_df = df[df['kategori'] == "OKSIJEN"]

# Düzeltilmiş veriyi gösterme
print(oksijen_df)
```

```python
# Model ve litre eşleşmeleri
# Tabloya göre litre eşleşmeleri
litre_values = {
    'OXY-GOLD-3': 300,
    'OXY-GOLD-7': 400,
    'OXY-GOLD-11': 1000,
    'OXY-GOLD-K15': np.nan,  # Değer yok
    'OXY-GOLD-15': 1000,
    'OXY-GOLD-18': 1000,
    'OXY-GOLD-22': 1000,
    'OXY-GOLD-30': 1500,
    'OXY-GOLD-37': 2000,
    'OXY-GOLD-45': 2000,
    'OXY-GOLD-55': 3000,
    'OXY-GOLD-75': 4000,
    'OXY-GOLD-90': 5000,
    'OXY-GOLD-110': 5000,
    'GOLDBERG-22-V': 1000,  
    'GOLDBERG-45-V': 2000   
}

# "OKSİJEN" kategorisindeki litre değerlerini güncelleme
for index, row in df.iterrows():
    if row['kategori'] == 'OKSIJEN':
        model = row['urun_modeli']
        print(f"Checking model: {model}, Current Litre: {row['litre']}")  # Kontrol için model ve mevcut litreyi yazdır
        if model in litre_values:
            new_value = litre_values[model]
            # NaN kontrolü yap
            if pd.notna(new_value):  
                print(f"Updating '{model}' to '{new_value}'")  # Güncelleme işlemini göster
                df.at[index, 'litre'] = new_value
            else:
                print(f"No valid mapping for '{model}' (NaN)")  # NaN durumu
        else:
            print(f"No mapping found for '{model}'")  # Eşleşme yoksa mesaj

# Sonuçları gösterme
print("\nGüncellenmiş DataFrame:")
print(df)
```
