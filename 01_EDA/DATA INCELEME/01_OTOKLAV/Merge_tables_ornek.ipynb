
Tabloları `pd.concat()` fonksiyonu ile birleştirmek için kullanılan bu kodu ve parametrelerini ayrıntılı olarak açıklayalım:

```python
pd.concat([table1, table2, table3, table4, table5], ignore_index=True, sort=False)
```

### Açıklama:
- **`pd.concat()`**: Bu fonksiyon, birden fazla DataFrame'i birleştirmek için kullanılır. Birleştirme işlemi satır veya sütun bazında yapılabilir.
- **`[table1, table2, table3, table4, table5]`**: Birleştirilecek DataFrame'ler bir liste içinde belirtilir.
- **`ignore_index=True`**: İndekslerin sıfırdan başlamasını sağlar ve her DataFrame'deki mevcut indeksleri göz ardı eder. Bu, özellikle farklı tablolardan gelen indekslerin karışıklığına engel olmak için kullanılır.
- **`sort=False`**: Varsayılan olarak Pandas, sütunları alfabetik sırayla yerleştirir, ancak `sort=False` ile sütunların orijinal sırasının korunmasını sağlar.

### Farklı Ayarlanabilecek Parametreler:

1. **`axis`**:
   - Birleştirmenin satır mı (varsayılan: `axis=0`), yoksa sütun bazında mı yapılacağını belirler.
   - **`axis=0`**: Satır bazında birleştirme (tabloyu alt alta ekleme). 
   - **`axis=1`**: Sütun bazında birleştirme (tabloyu yan yana ekleme).
   ```python
   pd.concat([table1, table2], axis=1)
   ```

2. **`join`**:
   - Sütun isimleri aynı olmayan DataFrame'lerde hangi birleşim türünün kullanılacağını belirtir.
   - **`join='outer'`**: Varsayılan değer olup, tüm sütunları içerir ve eksik verileri `NaN` ile doldurur.
   - **`join='inner'`**: Sadece her iki DataFrame'de de bulunan ortak sütunları içerir.
   ```python
   pd.concat([table1, table2], join='inner')
   ```

3. **`keys`**:
   - Birleştirilen her DataFrame'e ayırt edici bir anahtar ekler. Bu, çoklu indeks yapıları oluşturmak için kullanılabilir.
   ```python
   pd.concat([table1, table2], keys=['table1', 'table2'])
   ```

4. **`verify_integrity`**:
   - Bu parametre, birleştirilen DataFrame'lerin indekslerinin çakışıp çakışmadığını kontrol eder. Eğer indeksler çakışıyorsa hata verir.
   ```python
   pd.concat([table1, table2], verify_integrity=True)
   ```

5. **`copy`**:
   - Bu parametre, birleştirilen DataFrame'lerin kopyalanıp kopyalanmayacağını kontrol eder. Varsayılan olarak `copy=True` şeklinde ayarlanmıştır, yani orijinal veriler kopyalanır.
   ```python
   pd.concat([table1, table2], copy=False)
   ```

### Örnek:
```python
# Farklı join türü ve axis kullanımı örneği
combined_df = pd.concat([table1, table2, table3], axis=0, join='outer', ignore_index=True, sort=False)
```

Bu fonksiyon, tablo birleştirme işlemlerinde esneklik sağlar ve çeşitli parametrelerle tablo yapınızı ihtiyaçlarınıza göre şekillendirebilirsiniz.
